## 1. Debugeo
Para poder evaluar si un cliente realizo un pago, se mira la próxima cuota morosa que presenta. Si con el avance de los días se paga la última cuota morosa, se toma la del próximo mes y los días de mora disminuyen. Por esto, si algún cliente tiene una disminución en los días de mora, se dice que tiene un pago.
Para esto se usa el siguiente ejemplo:

```python
# Definir contenido de la tabla.
Cuenta = ["A","A","A","A","A","A","B","B","B","B","B","B","D","D","D","D","D","D","C","C","C","C","C","C"]
Fecha_proximo_vencimiento = ["07/09/2022","07/09/2022","07/09/2022","07/09/2022","07/09/2022","07/09/2022","20/09/2022","20/09/2022","20/09/2022","20/10/2022","20/10/2022","20/10/2022","11/10/2022","11/10/2022","11/10/2022","11/10/2022","11/10/2022","11/10/2022","07/07/2022","07/07/2022","07/07/2022","07/07/2022","07/07/2022","07/08/2022"]
Fecha_actual = ["03/10/2022","04/10/2022","05/10/2022","06/10/2022","11/10/2022","12/10/2022", "03/10/2022","04/10/2022","05/10/2022","06/10/2022","11/10/2022","12/10/2022","03/10/2022","04/10/2022","05/10/2022","06/10/2022","11/10/2022","12/10/2022","03/10/2022","04/10/2022","05/10/2022","06/10/2022","11/10/2022","12/10/2022"]
Dias_de_mora = [26, 27, 28, 29, 34, 35, 13, 14, 15, 0, 0, 0, 0, 0, 0, 0, 0, 1, 88, 89, 90, 91, 96, 66]
Dia_habil = [1,2,3,4,6,7,1,2,3,4,6,7,1,2,3,4,6,7,1,2,3,4,6,7]

# Creacion del dataframe.
Tabla = pd.DataFrame({"Cuenta" : Cuenta,
                        "Fecha proximo vencimiento": Fecha_proximo_vencimiento,
                        "Fecha Actual" : Fecha_actual,
                        "Dias de mora" : Dias_de_mora,
                        "Dia habil" : Dia_habil
                         })
#Cambio de formato. Transformar fechas de tipo String a Datetime. 
Tabla['Fecha proximo vencimiento'] = pd.to_datetime(Tabla['Fecha proximo vencimiento'], format="%d/%m/%Y")
Tabla['Fecha Actual'] = pd.to_datetime(Tabla['Fecha Actual'], format="%d/%m/%Y")

# Generar el dataframe Tabla2 a partir de la copia de Tabla. Al definir deep=True se asegura la creacion de un nuevo set de datos.
Tabla2 = Tabla.copy(deep=True)
Tabla2['Dia habil'] +=1

#Cambiar nombre a las columnas.
Tabla2.rename(columns={"Fecha proximo vencimiento" : "Fecha proximo vencimiento2",
                "Fecha Actual" : "Fecha Actual2",
                "Dias de mora" : "Dias de mora2"}, inplace=True)
Tabla2
```

## Visualizacion de dias morosos.

```python
l_chart = px.line(Tabla,
                  x='Fecha Actual', 
                  y='Dias de mora',
                  color='Cuenta',
                  hover_name='Cuenta',
                  title='Variacion de morosidad por cuenta a traves del tiempo.',)
 
l_chart.update_layout(xaxis_title='Dias',
                      yaxis_title='Dias de mora')
 
l_chart.show()
```

```python
#Unir las tablas creadas en base a las columnas "Cuenta" y "Dia habil".
Tabla3 = pd.merge(Tabla,Tabla2, on=("Cuenta","Dia habil") ,how='left')
#Creacion de columna "Pago". Aquellas cuentas que presentan un aumento en los dias de mora, se registran con uno, de lo contrario, se inserta cero.
Tabla3['Pago'] = np.where(Tabla3['Dias de mora2'] > Tabla3['Dias de mora'], 1, 0)
Tabla3.head()
```

```python
#Agrupacion registros en funcion de cuentas. La columna "Pago" contendra la suma de todas las filas de la hilera.
Tabla4 =  Tabla3.groupby(['Cuenta'], as_index=False).agg({'Pago': pd.Series.sum})
#Se agrega una columna "Tiene_pago". Cuentas con pagos mayores a 0 se registraran "PAGO", lo opuesto sera "NO PAGO".
Tabla4['TIENE_PAGO'] = np.where(Tabla4['Pago'] > 0, "PAGO", "NO PAGO")
Tabla5 = Tabla4[["Cuenta","TIENE_PAGO"]].copy() # Creacion de nueva tabla con copia columnas "Cuenta","TIENE_PAGO"
Tabla5
```
